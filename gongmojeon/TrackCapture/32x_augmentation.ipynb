```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
oripath = "./TrackCapture/TrackCapture_original"
fnames = os.listdir(oripath)

oriimgs = [] # 원본 이미지
orinums = [] # [orinum]_angle.png
angles = [] # orinum_[angle].png

for fname in fnames:
  oriimg_big = cv2.imread(oripath + '/' + fname)
  oriimg = cv2.resize(oriimg_big, (64,48))
  plt.imshow(oriimg)
  oriimgs.append(np.array(oriimg))
  orinums.append(int(fname[0:2]))
  angles.append(int(fname[3:5]))

oriimgs = np.array(oriimgs)
orinums = np.array(orinums)
angles = np.array(angles)
```

```python
# augment 후 : (38, 54, 3)
allaugimgs = []
for oriimg in oriimgs:
  augimgs = []
  for i in range(10):
    for j in range(10):
      ii = oriimg[i:i+38,j:j+54]
      augimgs.append(ii)
  allaugimgs.append(np.array(augimgs))
```

```python
angidx = [np.where(angles==90)[0], np.where(angles==30)[0]]

augimgs90 = []
augimgs30 = []

for idx90 in angidx[0]:
  for ii in allaugimgs[idx90]:
    augimgs90.append(ii)
for idx30 in angidx[1]:
  for ii in allaugimgs[idx30]:
    augimgs30.append(ii)

random.shuffle(augimgs90)
random.shuffle(augimgs30)

augimgs90 = np.array(augimgs90)
augimgs30 = np.array(augimgs30)
```

```python
finaugimgs = np.concatenate([augimgs90, augimgs30])
finlabels = np.concatenate([90*np.ones(len(augimgs90)), 30*np.ones(len(augimgs30))])
print(len(finaugimgs))
print(len(finlabels))
```

```python
totnum = 0
for idx in range(len(finaugimgs)):
  imgnum = str('0' * (4-len(str(totnum))) + str(totnum))
  fname = './TrackCapture/TrackCapture_augmentation/' + 'Train_' + imgnum + '_' + str(int(finlabels[idx])) + '.png'
  cv2.imwrite(fname, finaugimgs[idx])
  totnum += 1
  print(totnum)
```
